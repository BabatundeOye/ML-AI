```python
#split the data into features and target variable
X = df.drop(columns = 'diagnosis').values
y = df['diagnosis'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python

model =MLPClassifier(hidden_layer_sizes = (80,50,25), 
                     activation = 'relu',
                     solver = 'adam',
                     batch_size = 20,
                     max_iter=270, 
                     random_state=42,
                     verbose=True,
                     early_stopping=True,
                     tol = 0.0001,
                     validation_fraction=0.1)
```

```python
model.fit(X_train, y_train)
```

```python
print("The model has a test data accuracy of:",model.score(X_test,y_test))
print("The model has a training data accuracy of:",model.score(X_train,y_train))
```

```python
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
#get important features
feature_importance = np.abs(model.coefs_[0]).sum(axis=1)
feature_importance = pd.Series(feature_importance, index=df.drop(columns='diagnosis').columns)
print(feature_importance)
```

```python

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping


#split the data into features and target variable
X = df.drop(columns = 'diagnosis').values
y = df['diagnosis'].values
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define the model
model = Sequential([
    Dense(80, activation='relu', input_shape=(17,)),  # input layer + first hidden layer
    Dense(50, activation='relu'),                       # second hidden layer
    Dense(25, activation='relu'),                       # third hidden layer
    Dense(1, activation='sigmoid')                      # output layer (binary classification)
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Early stopping callback
early_stop = EarlyStopping(monitor='val_loss',
                           patience=10,            # similar to scikit-learn early_stopping
                           restore_best_weights=True,
                           min_delta=0.0001)

# Train the model
history = model.fit(X_train, y_train,
                    epochs=270,
                    batch_size=20,
                    validation_split=0.1,
                    verbose=1,
                    callbacks=[early_stop])
```
