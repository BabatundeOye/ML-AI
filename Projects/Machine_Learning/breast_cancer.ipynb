```python
# Calculate the correlation matrix
correlation_matrix = df_without_first_column.corr()

# Set a correlation threshold (e.g., 0.85 for high correlations)
threshold = 0.85

# Create a mask for the upper triangle of the correlation matrix (to avoid duplication)
mask = (correlation_matrix.abs() > threshold) & (correlation_matrix != 1)

#abs 
#The absolute value of 5 is 5.
#The absolute value of -5 is 5.

# Mask the correlation matrix to remove highly correlated values
correlation_matrix_filtered = correlation_matrix.where(~mask)
#~mask: 
# The tilde (~) is a logical NOT operator. It inverts the boolean values in the mask.
# So, if the mask was True, it becomes False, and if it was False, it becomes True. 
# Essentially,it selects the correlations that are below the threshold and not self-correlated.
# Display the filtered correlation matrix


#.where~mask
#This function returns a new matrix where it keeps the values where ~mask is True 
#(those that are below the threshold and not self-correlated). For the places where 
#~mask is False (correlations above the threshold or self-correlated), 
#the result will be NaN (Not a Number).

plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix_filtered, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title for the heatmap
plt.title(f"Correlation Matrix (Filtered, Threshold={threshold})")

# Show the plot
plt.show()

plt.savefig("correlation_matrix.pdf", format="pdf", bbox_inches="tight")
```

```python

# Set a correlation threshold 
threshold = 0.85

# Find the pairs of features where the absolute correlation is above the threshold
high_corr_pairs = []

# Loop over the upper triangle of the correlation matrix (to avoid duplication)
#it compares 0 with the whole dataset, then 1, then 2 ........
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold: #starts from j=0, i=1
            high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j],
                                     correlation_matrix.iloc[i, j]))

# Print the pairs of features with high correlation
print("Highly Correlated Pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}")
```
